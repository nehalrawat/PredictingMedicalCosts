```julia


Other = 0
sumOther = 0
for x = 1:length(BronxData[:, :Race])
    if ((BronxData[x, :Race] == "Other Race"))
        Other = Other + BronxData[x, :Total_Charges]
        sumOther = sumOther + 1
    end
end

avgOther = Other/sumOther
println(avgOther)

Black = 0
sumBlack = 0
for x = 1:length(BronxData[:, :Race])
    if ((BronxData[x, :Race] == "Black/African American"))
        Black = Black + BronxData[x, :Total_Charges]
        sumBlack = sumBlack + 1
    end
end

avgBlack = Black/sumBlack
println(avgBlack)

White = 0
sumWhite = 0
for x = 1:length(BronxData[:, :Race])
    if ((BronxData[x, :Race] == "White"))
        White = White + BronxData[x, :Total_Charges]
        sumWhite = sumWhite + 1
    end
end

avgWhite = White/sumWhite
println(avgWhite)


Multi = 0
sumMulti = 0
for x = 1:length(BronxData[:, :Race])
    if ((BronxData[x, :Race] == "Multi"))
        Multi = Multi + BronxData[x, :Total_Charges]
        sumMulti = sumMulti + 1
    end
end

avgMulti = Multi/sumMulti
println(sumMulti)
println(avgMulti)


Unknown = 0
sumUnknown = 0
for x = 1:length(BronxData[:, :Race])
    if ((BronxData[x, :Race] == "U"))
        Unknown = Unknown + BronxData[x, :Total_Charges]
        sumUnknown = sumUnknown + 1
    end
end

avgUnknown = Unknown/sumUnknown
println(avgUnknown)
println(sumUnknown)



  
```

```julia
bar(["White","Black","Other"], [avgWhite, avgBlack, avgOther])
xlabel("Race")
ylabel("Average Hospital Charge")
title("Average Bronx Hospital Charges for Chest Pain vs. Race")
```
